```python
import random

from hypex import ABTest, SQLiteTracker
from hypex.ui.base import ExperimentShell
from hypex.dataset import Dataset, InfoRole, TargetRole, TreatmentRole
from hypex.utils import create_test_data
```

```python
df=create_test_data()
df["treat"] = [random.choice([0, 1, 2]) for _ in range(len(df))]
data = Dataset(
    roles={
        "user_id": InfoRole(int),
        "treat": TreatmentRole(),
        "pre_spends": TargetRole(),
        "post_spends": TargetRole(),
        "gender": TargetRole()
    }, data=df,
)
data
```

```python

tracker = SQLiteTracker(db_path="./experiment_history.db")

ab_test = ABTest(
    additional_tests=["t-test", "chi2-test"],
    multitest_method="holm"
)

shell = ExperimentShell(
    experiment=ab_test.experiment,
    output=ab_test._out,
    tracker=tracker
)

results = shell.execute(data)

print(results.resume)
```
